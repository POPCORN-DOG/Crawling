```python
def search(search_word,page_num=1,page_size=40,cross_border_shopping='false',mall='',catid=''):
    try:
        dic = {
            '쿠팡':'613264',
            '위메프':'197023',
            '티몬':'221844',
            'G9':'280455',
            '이마트':'596',
            '롯데마트':'107396',
            '홈플러스':'108756',
            '롯데프레시':'235018',
            '인터파크':'3',
            '11번가':'17703',
            '':''
        }

        dic2 = {
            '백미':'50004749',
            '현미':'50004751',
            '':''
        }
        
        url = 'https://search.shopping.naver.com/search/all?' +\
                'catId=' + dic2[catid] +\
                '&exagency=' + cross_border_shopping +\
                '&frm=NVSHATC&origQuery='+ search_word +\
                '&pagingIndex='+ str(page_num) +\
                '&pagingSize='+ str(page_size) +\
                '&productSet=total&query='+ search_word +\
                '&sort=rel&timestamp=&viewType=list'+\
                '&mall=' + dic[mall]
        return(url)
    
    except:
        print('잘못입력하였습니다.')
```

```python
def soup2select(url):
    headers = {
    'User-Agent':'Mozilla/5.0 (Windows NT 6.1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/41.0.2228.0 Safari/537.36',
    }
    
    response = requests.get(url,headers=headers,verify=False)
    soup     = BeautifulSoup(response.text,'lxml')
    soup2    = soup.select('body > script')[0].text.replace('window.__PRELOADED_STATE__=','')
    return(soup2)
```

```python
def data(soup2select):
    soup2_j  = json.loads(soup2select)
    soup2_li = soup2_j['props']['pageProps']['initialState']['products']['list']
    return(soup2_li)
```

```python
productTitle = [] # 겉 제목
for i in tnrange(100):
    tt = data(soup2select(search('현미 -쌀눈',i,80,'true','','현미')))
    for i in range(len(tt)):
        productTitle.append(tt[i]['item']['productTitle'])
```
