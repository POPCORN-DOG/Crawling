```python
def search(search_word,page_num=1,page_size=40,cross_border_shopping='false',mall=''):
    try:
        dic = {
            '쿠팡':'613264',
            '위메프':'197023',
            '티몬':'221844',
            'G9':'280455',
            '이마트':'596',
            '롯데마트':'107396',
            '홈플러스':'108756',
            '롯데프레시':'235018',
            '':''
        }
        
        url = 'https://search.shopping.naver.com/search/all?' +\
                'exagency=' + cross_border_shopping +\
                '&frm=NVSHATC&origQuery='+ search_word +\
                '&pagingIndex='+ str(page_num) +\
                '&pagingSize='+ str(page_size) +\
                '&productSet=total&query='+ search_word +\
                '&sort=rel&timestamp=&viewType=list'+\
                '&mall=' + dic[mall]
        return(url)
    
    except:
        print('잘못입력하였습니다.')
```

```python
def soup2select(url):
    headers = {
    'User-Agent':'Mozilla/5.0 (Windows NT 6.1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/41.0.2228.0 Safari/537.36',
    }
    
    response = requests.get(url,headers=headers,verify=False)
    soup     = BeautifulSoup(response.text,'lxml')
    soup2    = soup.select('body > script')[0].text.replace('window.__PRELOADED_STATE__=','')
    return(soup2)
```

```python
def final(search_word,page_num=1,page_size=40,cross_border_shopping='false',mall=''):
    s = search(search_word,page_num,page_size,cross_border_shopping,mall)

    tt = data(soup2select(s))

    productTitle = [] # 겉 제목
    productName = [] # 속 제목
    name = [] # 판매자
    adsrType = []
    url = []

    for i in tnrange(len(tt)):
        try:
            name.append(tt[i]['item']['mallInfoCache']['name'])
            productTitle.append(tt[i]['item']['productTitle'])
            productName.append(tt[i]['item']['productName'])
            adsrType.append(tt[i]['item']['mallInfoCache']['adsrType'])
            try:
                url.append(tt[i]['item']['mallProductUrl'])
            except:
                try:
                    url.append(tt[i]['item']['adcrUrl'])
                except:
                    url.append(tt[i]['item']['crUrl'])  
                    
        # 묶음 상품일 경우
        except KeyError : 
            print(i)
            for j in range(len(tt[i]['item']['lowMallList'])):
                name.append(tt[i]['item']['lowMallList'][j]['name'])
                productTitle.append(tt[i]['item']['productTitle'])
                productName.append('묶음')

                
                if tt[i]['item']['lowMallList'][j]['chnlSeq'] != '':
                    adsrType.append('SHOPN')
                else:
                    adsrType.append('CP')
                    
#                url.append(tt[i]['item']['crUrl']) 
                try:
                    url2 = 'https://search.shopping.naver.com/search/rd' +\
                            '?rank=1' +\
                            '&pagingIndex=1' +\
                            '&pagingSize=40' +\
                            '&query=' + search_word +\
                            '&bizCd=0301' +\
                            '&prntExpsTrtrCd=000073' +\
                            '&cpcExtrtrCd=000075' +\
                            '&nvMid=' + tt[i]['item']['lowMallList'][j]['nvMid'] +\
                            '&nclickArea=lst*C&naId=' +\
                            '&rankCatalog=1' +\
                            '&cntCatalog=5'

                    headers = {
                    #    'Accept':'application/json, text/plain, */*',
                    #    'Accept-Encoding':'gzip, deflate, br',
                    #    'Accept-Language':'ko,en;q=0.9,en-US;q=0.8',
                    #    'Connection': 'keep-alive',
                    #    'Referer':'https://search.shopping.naver.com/search/all?query=%EB%A7%88%EC%8A%A4%ED%81%AC&cat_id=&frm=NVSHATC',
                    #    'Use-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/85.0.4183.121 Safari/537.36',
                    #    'Host':'search.shopping.naver.com',
                    #    'Sec-Fetch-Dest':'empty',
                    #    'Sec-Fetch-Mode':'cors',
                    #    'Sec-Fetch-Site':'same-origin',
                        'urlprefix':'/api'
                    }

                    resp = requests.get(url2,headers=headers,verify=False)
                    url.append(resp.json()['redirect'])
                except KeyError as e:
                    print(e)
                    print(i,j)
                    url.append('수집실패')
                    continue
                    


    return(pd.DataFrame(
        {
            "productTitle":productTitle,
            "productName":productName,
            "adsrType":adsrType,
            "name":name,
            "url":url
        }
    ))
    
```
